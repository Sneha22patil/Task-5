```python
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Survived'], kde=True)
plt.title('Survival Status')
plt.show()
```

Observation:

The histogram shows the distribution of the Survived column, which is binary:

0 = Did not survive

1 = Survived

The majority of passengers did not survive (0 has a higher bar).

The plot highlights the imbalance in survival outcomes — more passengers perished than survived.

The KDE curve is not very meaningful here since Survived is categorical (discrete), not continuous — a count plot or bar plot would be more appropriate.

```python
sns.boxplot(y=df['Age'])
plt.title('Age Distibution')
plt.show()
```

Observation:

The boxplot shows the spread and central tendency of the age variable.

Most passengers were between approximately 20 and 40 years old (the interquartile range).

The median age lies around 28–30 years.

There are several outliers below 10 and above 60, including infants and elderly passengers.

This distribution suggests a younger majority on board, with a few much older or younger individuals.

```python
sns.scatterplot(x=df['Age'], y=df['Fare'], hue=df['Survived'])
plt.title("Age vs Fare by Survival")
plt.show()
```

Observation:

The scatterplot visualizes the relationship between passenger age and fare paid, colored by survival status.

There is no strong linear relationship between age and fare.

Survivors (hue=1) are more concentrated among:

Higher fares (many survivors paid expensive tickets, likely 1st class).

Younger passengers, including children and young adults.

A dense cluster of passengers paid low fares, mostly non-survivors, indicating that lower class (and fare) correlated with lower survival.

Some elderly passengers who paid high fares also survived — likely in 1st class.

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle("Pairplot Titanic Features", y=1.02, fontsize=14)
plt.show()
```

Observation:

The pairplot shows pairwise relationships between age, fare, pclass, and survived, colored by survival status.

Survivors (hue=1) tend to cluster around:

Lower pclass values (1st class),

Higher fares,

A wider range of ages, but more in younger to middle-age groups.

Non-survivors (hue=0) are mostly in:

Higher pclass values (3rd class),

Lower fares,

Similar age distribution but with less survival overall.

The diagonal plots show the distribution of each variable for survivors vs. non-survivors.

This visualization confirms that passenger class and fare are strong predictors of survival, while age shows a less distinct separation.

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title('Correlation Heatmap of Titanic Features')
plt.show()
```

Observation:

The heatmap displays the correlation coefficients between numeric features in the Titanic dataset.

Notable correlations:

Negative correlation between pclass and fare (~ -0.55): higher class (1st) passengers pay higher fares (since pclass is coded as 1 = 1st class, 3 = 3rd class).

Positive correlation between fare and survived (~ 0.26): passengers who paid higher fares had higher chances of survival.

Negative correlation between pclass and survived (~ -0.34): lower class passengers had lower survival rates.

Other numeric features like age show very weak correlations with survival.
